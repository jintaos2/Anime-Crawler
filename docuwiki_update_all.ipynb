```python
import requests
import re
import os
```

```python
def update_list(year):
    try:
        r=requests.get("https://docuwiki.net/index.php?title=Category:{}".format(year))
    except:
        print("error!")
        return
    path = 'docu/{}'.format(year)
    if not os.path.exists(path):
        os.mkdir(path)
    table = re.findall(r'<table[\s\S]*/table>',r.text)[0]
    table = re.sub(r'[\n\t]','',table)
    items = re.findall(r'<li>(.*?)</li>',table) 
    prev_list = read_prev_list(year)
    print("year: {} n_prev: {} \tn_new: {}".format(year, len(prev_list), len(items)))
    
    n = 1
    for i in items:
        release_title = re.findall(r'<a.*?>(.*?)</a>',i)[0]
        release_link = 'https://docuwiki.net'+re.findall(r'href="([^"]*)"', i)[0]
        rr = requests.get(release_link)
        release_content = re.sub(r'[\n\t]','',rr.text)
        save_one(release_title, release_content, prev_list, year)
        print("updating {}\t of {} items".format(n,len(items)))
        n+=1
    
def read_prev_list(year):
    file = 'docu/{}/all.txt'.format(year)
    if not os.path.exists(file):
        return []
    with open(file, 'r', encoding='utf8') as f:
        content = f.read()
    prev_list = re.findall(r'<p>[\s](.*)', content)
    f = open(file, 'w', encoding='utf8').close()
    return prev_list
    
    
def save_one(release_title, release_content, prev_list, year):
    summary_ = re.findall(r"<p>.*?published by.*?</p>",release_content)
    if len(summary_) < 1:
        summary_ = re.findall(r'<p>.*?</p>',release_content)
    summary = re.sub(r'<.*?>','',summary_[0])
    ed2ks_raw = re.findall(r"<a[^>]*?ed2klink[^>]*?>.*?</b>",release_content)
    ed2ks = []
    for i in ed2ks_raw:
        ed2k_title = re.sub(r'<.*?>','',i)
        ed2k_title = re.sub(r'&nbsp;',' ',ed2k_title)
        ed2k_link = re.findall(r'href="([^"]*)"',i)[0] 
        ed2ks.append([ed2k_title, ed2k_link])
    file1 = 'docu/{}/all.txt'.format(year)
    file2 = 'docu/{}/latest.txt'.format(year)
    write_one(file1, release_title, summary, ed2ks)
    if not release_title in prev_list:
        write_one(file2, release_title, summary, ed2ks)
    
    
def write_one(filename, release_title, summary, ed2ks):
    with open(filename, 'a+', encoding='utf8') as f:
        f.write("<p>\n{}\n{}\n".format(release_title, summary))
        for i in ed2ks:
            f.write("{},{},\n".format(i[0],i[1]))
        f.write("</p>\n\n")
```

```python
update_list(2001)
```
